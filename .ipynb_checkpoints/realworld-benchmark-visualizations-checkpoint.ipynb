```python
#Dataframe extraction

#real dataset cojac results
df_coj_real =pd.read_csv("data/real/cojac_real.tsv", sep='\t').fillna(0)
df_coj_real_ca = df_coj_real.loc[df_coj_real['Location'] == 'Canada']
df_coj_real_us = df_coj_real.loc[df_coj_real['Location'] == 'US']
```

# Real worls datasets: Canada, Ontario (PRJNA824537), US (PRJNA765346)

To look at Cojac results for real world datasets, barplots similar to Freyja barplots were generated using Python matplotlib library.

#### Visualization of Cojac results for Canada, Ontario (PRJNA824537) dataset

```python
# set the figure size
plt.figure(figsize=(7.5, 10))

# from raw value to percentage
total = df_coj_real_ca.groupby('Date')['Proportion'].sum().reset_index()
alpha = df_coj_real_ca[df_coj_real_ca.Lineage=='Alpha'].groupby('Date')['Proportion'].sum().reset_index()
delta = df_coj_real_ca[df_coj_real_ca.Lineage=='Delta'].groupby('Date')['Proportion'].sum().reset_index()
ba1 = df_coj_real_ca[df_coj_real_ca.Lineage=='Omicron (BA.1)'].groupby('Date')['Proportion'].sum().reset_index()
ba2 = df_coj_real_ca[df_coj_real_ca.Lineage=='Omicron (BA.2)'].groupby('Date')['Proportion'].sum().reset_index()
other = df_coj_real_ca[df_coj_real_ca.Lineage=='Other'].groupby('Date')['Proportion'].sum().reset_index()

alpha['Proportion'] = [i / j for i,j in zip(alpha['Proportion'], total['Proportion'])]
delta['Proportion'] = alpha['Proportion'] + delta['Proportion']
ba1['Proportion'] =delta['Proportion'] + ba1['Proportion']
ba2['Proportion'] = ba1['Proportion'] + ba2['Proportion']
total['Proportion'] = ba2['Proportion'] + other['Proportion']

# bar charts
bar1 = sns.barplot(x="Date", y="Proportion", data=total, color='#aec6e8', order=['Nov21', 'Dec21', 'Jan22','Feb22'])
bar2 = sns.barplot(x="Date", y="Proportion", data=ba2, color='#f8bc83', order=['Nov21', 'Dec21', 'Jan22','Feb22'])
bar4 = sns.barplot(x="Date", y="Proportion", data=ba1, color='#1f78b4', order=['Nov21', 'Dec21', 'Jan22','Feb22'])
bar6 = sns.barplot(x="Date", y="Proportion", data=delta, color='#fa7d11', order=['Nov21', 'Dec21', 'Jan22','Feb22'])
bar7 = sns.barplot(x="Date", y="Proportion", data=alpha, color='#2aa129', order=['Nov21', 'Dec21', 'Jan22','Feb22'])

# add legend
bar_floor5 = mpatches.Patch(color='#aec6e8', label='Other')
bar_floor4 = mpatches.Patch(color='#f8bc83', label='Omicron (BA.2)')
bar_floor3 = mpatches.Patch(color='#1f78b4', label='Omicron (BA.1)')
bar_floor2 = mpatches.Patch(color='#fa7d11', label='Delta')
bar_floor1 = mpatches.Patch(color='#2aa129', label='Alpha')
plt.legend(handles=[bar_floor5, bar_floor4, bar_floor3, bar_floor2, bar_floor1], loc='upper left', bbox_to_anchor=(1.05, 1))
plt.title(label='Cojac results on dataset from Canada, Ontario (PRJNA824537)')

# show the graph
plt.show()
```
